## EDA

```python
#convert date_of_journey to timestamp
train_data['Journey_day'] = pd.to_datetime(train_data.Date_of_Journey, format="%d/%m/%Y").dt.day
train_data['Journey_month'] = pd.to_datetime(train_data.Date_of_Journey, format="%d/%m/%Y").dt.month
```

```python
train_data.head()
```

```python
train_data.drop(['Date_of_Journey'], axis=1, inplace=True)
```

```python
#extracting values from Dep_Time

train_data['Dep_hour'] = pd.to_datetime(train_data.Dep_Time).dt.hour
train_data['Dep_min'] = pd.to_datetime(train_data.Dep_Time).dt.minute

train_data.drop(['Dep_Time'], axis=1, inplace=True)

train_data.head()
```

```python
#extracting values from Arrival_Time

train_data['Arrival_hour'] = pd.to_datetime(train_data.Arrival_Time).dt.hour
train_data['Arrival_min'] = pd.to_datetime(train_data.Arrival_Time).dt.minute

train_data.drop(['Arrival_Time'], axis=1, inplace=True)

train_data.head()
```

```python
#pre-processing the duration column

duration = list(train_data['Duration'])

for i in range(len(duration)):
    if len(duration[i].split()) != 2:          #checking if it contains hr or mins also
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"     #adds 0 minutes
        else:
            duration[i] = "0h " + duration[i]
            
duration_hours = []
duration_mins = []

for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0]))     #extract hr
    duration_mins.append(int(duration[i].split(sep = "m")[0].split()[-1]))     #extract only mins
```

```python
train_data['Duration_hours'] = duration_hours
train_data['Duration_mins'] = duration_mins
train_data.head()
```

```python
train_data.drop(['Duration'], axis=1, inplace=True)
train_data.head()
```

```python
#OHE on Source column

source = train_data[['Source']]
source = pd.get_dummies(source, drop_first=True)

source.head()
```

```python
train_data['Destination'].value_counts()
```

```python
#OHE on Destination column

destination = train_data[['Destination']]
destination = pd.get_dummies(destination, drop_first=True)

destination.head()
```

Route and Total_Stops has similar information.

```python
#concatenate all the encoded values in our dataframe

data_train = pd.concat([train_data, airline, source, destination], axis=1)

data_train.head()
```

## Hyperparameter Tuning

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
n_estimators = [int(x) for x in np.linspace(start=100, stop=1200, num=12)]

max_features = ['auto', 'sqrt']     #no. of features at each split

max_depth = [int(x) for x in np.linspace(5, 30, num=6)]     #max levels in tree

min_samples_split = [2,5,10,15,100]     #min samples req to split a node

min_samples_leaf = [1,2,5,10]     #min samples req at each leaf node
```

```python
#random grid

random_grid = {'n_estimators': n_estimators,
              'max_features': max_features,
              'max_depth' : max_depth,
              'min_samples_leaf' : min_samples_leaf,
              'min_samples_split' : min_samples_split}
```

```python
#random searching parameters using 5 fold CV, search across 100 diff combos
rf_random = RandomizedSearchCV(estimator= reg_rf, param_distributions=random_grid, scoring='neg_mean_squared_error', n_iter=10, cv=5, verbose=2, random_state=52, n_jobs=1)
```

```python
rf_random.fit(X_train, y_train)
```

```python
rf_random.best_params_
```

```python
prediction = rf_random.predict(X_test)
```

```python
plt.figure(figsize=(8,8))
sns.distplot(y_test - prediction)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.scatter(y_test, prediction, alpha=0.5)
plt.xlabel("y_test")
plt.ylabel("prediction")
plt.show()
```

```python
print("MAE: ", metrics.mean_absolute_error(y_test, prediction))
print("MSE: ", metrics.mean_squared_error(y_test, prediction))
print("RMSE: ", np.sqrt(metrics.mean_squared_error(y_test, prediction)))
```
